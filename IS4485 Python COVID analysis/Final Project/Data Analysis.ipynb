# Bar Charts

### Top 10 Internationally Influential Countries - Totals

```python
plt.style.use('ggplot')

# Setting width of bars
w = 0.1

# Setting x labels
x = ["US", "UK", "China", "Russia", "Germany", 
    "France", "Japan", "Italy", "Emirates", 
    "Israel"]

# Setting bar values
country_lst = ["US", "United Kingdom", "China", "Russia", "Germany", 
    "France", "Japan", "Italy", "United Arab Emirates", 
    "Israel"]
country_lst_vax = ["United States", "United Kingdom", "China", "Russia", "Germany", 
    "France", "Japan", "Italy", "United Arab Emirates", 
    "Israel"]
deaths = da.get_total_lst(False, country_lst, df_dths)
confirmed = da.get_total_lst(False, country_lst, df_conf)
recovered = da.get_total_lst(False, country_lst, df_reco)
dailyvax = da.get_total_lst(True, country_lst_vax, df_dailyvax)

# Creating bar clusters
bar1 = np.arange(len(x))
bar2 = [i + w for i in bar1]
bar3 = [i + w for i in bar2]
bar4 = [i + w for i in bar3]

# Plotting bars
plt.bar(bar1, deaths, w, label="Deaths")
plt.bar(bar2, confirmed, w, label="Confirmed")
plt.bar(bar3, recovered, w, label="Recovered")
plt.bar(bar4, dailyvax, w, label="Vaccinated")

# Creating multiple bar chart
plt.style.use('bmh')
plt.xlabel("Countries")
plt.ylabel("Totals")
plt.title("Top 10 Internationally Influential Countries Totals\n")

bar_ticks = [i + w/2 for i in bar2]
plt.xticks(bar_ticks, x, rotation='vertical')

plt.legend()
plt.show()
```

##### Notes: The number of people vaccinated is way higher than the number of people confirmed. This shows that there are many people who are vaccinated who have not caught COVID.
